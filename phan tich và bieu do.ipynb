

### Subtask:
Phân tích nội dung của cột 'utility'. Có thể cần trích xuất thông tin cụ thể như RAM, dung lượng lưu trữ, loại chip, v.v.


**Reasoning**:
Display the first few entries of the 'utility' column to understand its structure and content.



```
display(df['utility'].head())
```

**Reasoning**:
Based on the displayed entries, the 'utility' column contains concatenated strings with specifications separated by newline characters. I can extract RAM, storage, and chip information by splitting the string by newline and looking for specific keywords like 'RAM:', 'Dung lượng:', and 'Chip'.



```
def extract_specs(utility_string):
    specs = {}
    if isinstance(utility_string, str):
        lines = utility_string.split('\n')
        for line in lines:
            if 'Chip' in line:
                specs['Chip'] = line.replace('Chip', '').strip()
            elif 'RAM:' in line:
                specs['RAM'] = line.replace('RAM:', '').strip()
            elif 'Dung lượng:' in line:
                specs['Storage'] = line.replace('Dung lượng:', '').strip()
    return specs

df['specs'] = df['utility'].apply(extract_specs)
display(df[['utility', 'specs']].head())
```

**Reasoning**:
The 'specs' column now contains dictionaries with extracted information. I can further process this column to create separate columns for 'Chip', 'RAM', and 'Storage' and then finish the subtask.



```
df['Chip'] = df['specs'].apply(lambda x: x.get('Chip'))
df['RAM'] = df['specs'].apply(lambda x: x.get('RAM'))
df['Storage'] = df['specs'].apply(lambda x: x.get('Storage'))

display(df[['Chip', 'RAM', 'Storage']].head())
```

Phân tích nội dung của cột 'description'. Có thể thực hiện phân tích văn bản cơ bản như đếm số từ phổ biến, hoặc phân loại mô tả.


**Reasoning**:
Display the first 5 rows of the 'description' column to understand its content and format.



```
display(df['description'].head())
```

**Reasoning**:
Perform word frequency analysis on the 'description' column and display the top 10 most frequent words.



```
from collections import Counter
import re

# Combine all descriptions into a single string
all_descriptions = ' '.join(df['description'].dropna().tolist())

# Remove punctuation and convert to lowercase
all_descriptions = re.sub(r'[^\w\s]', '', all_descriptions.lower())

# Split the string into words
words = all_descriptions.split()

# Count word frequencies
word_counts = Counter(words)

# Display the 10 most common words
display(word_counts.most_common(10))
```

## Kết hợp thông tin

### Subtask:
Kết hợp thông tin từ các cột khác nhau để tạo ra các phân tích sâu hơn (ví dụ: so sánh giá dựa trên RAM hoặc loại chip).


**Reasoning**:
Group the DataFrame by the 'RAM' column and calculate the average 'price' for each RAM category, then group by the 'Chip' column and calculate the average 'price'.



```
avg_price_by_ram = df.groupby('RAM')['price'].mean().sort_values(ascending=False)
display(avg_price_by_ram)

avg_price_by_chip = df.groupby('Chip')['price'].mean().sort_values(ascending=False)
display(avg_price_by_chip)
```

**Reasoning**:
Visualize the average price by RAM and Chip using bar plots for better understanding.



```
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))
sns.barplot(x=avg_price_by_ram.index, y=avg_price_by_ram.values, palette='viridis')
plt.title('Average Price by RAM')
plt.xlabel('RAM')
plt.ylabel('Average Price (VND)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

plt.figure(figsize=(15, 7))
sns.barplot(x=avg_price_by_chip.index, y=avg_price_by_chip.values, palette='plasma')
plt.title('Average Price by Chip')
plt.xlabel('Chip')
plt.ylabel('Average Price (VND)')
plt.xticks(rotation=90, ha='right')
plt.tight_layout()
plt.show()
```
